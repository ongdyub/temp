```python
# TXT to JSON

import json

# Path to your text file
file_path = '/workspace/exp/chord_group_2.txt'
file_path = '/workspace/data/corpus/rhythm_token_cnt.txt'
# Initialize an empty dictionary to store the processed data
result_dict = {}

# Open the file and process each line
with open(file_path, 'r') as file:
    for idx, line in enumerate(file):
        if idx == 0:
            continue
        # Split the line into the code part and the number part
        code, number = line.strip().split()
        # Remove the last character from the code part to shorten it
        # Add to the dictionary with a value of 1
        result_dict[idx+2] = code
    result_dict[0] = '<pad>'
    result_dict[1] = '<eos>'
    result_dict[2] = '<bos>'
# Convert the dictionary to a JSON string
json_result = json.dumps(result_dict, indent=4)

# Optionally, save the JSON to a file or print it
print(json_result)

# Save the JSON result to a file
with open('/workspace/data/vocabs/rhythm_token.json', 'w') as json_file:
    json_file.write(json_result)

```
